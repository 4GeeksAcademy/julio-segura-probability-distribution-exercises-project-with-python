**Binomial Distribution**

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either Head or Tails.

```python
# Probability functions, density and distribution



# Calculate the point estimate of the binomial distribution
'''Esta función usa la función binom.pmf() para calcular la funcuón de masa de probabilidad
de una distribución binomial. La PMF devuelve la probabilidad de obtener un número específico de éxitos (x)
en una muestra de tamaño determinado (size) con una probabilidad de éxito (prob) dada'''

def dbinom(x,size,prob=0.5): 
    # x = el número de éxitos que queremos evaluar
    # size = el nº total de ensayos o eventos en la distribución
    # prob = la probabilidad de éxito en cada ensayo

    from scipy.stats import binom

    result = binom.pmf(k=x,n=size,p=prob,loc=0) # k = x = exitos
    # Calcula la PMF de la distribución binomial utilizando la función binom.pmf(). 
    # Esta función toma varios argumentos: k (número de éxitos), n (tamaño de muestra), p (probabilidad de éxito) 
    # y loc (desplazamiento, en este caso se establece en 0). 

    # El desplazamiento (loc = 0) hace referencia al ajuste desde donde se comienzan a contar los éxitos. 
    # En este caso, se cuentan desde 0
   
    return result


# Calculate the cumulative of the binomial distribution
'''Esta función usa la función binom.cfd() para calcular la función de distribución acumulada 
de una distribución binomial. La CDF devuelve la probabilidad acumulada de obtener un cierto número de 
éxitos (q), en una muestra de un tamaño determinado (size), con una probabilidad de éxito (prob) dada.'''

def pbinom(q,size,prob=0.5): # q = exitos
    
    from scipy.stats import binom
    
    result=binom.cdf(k=q,n=size,p=prob,loc=0)
    # binom.cfd() calcula la CDF de la distribución binomial. Esta función toma varios argumentos: 
    # k (número de éxitos hasta el cual se desea calcular la CDF), 
    # n (tamaño de la muestra que se va a calcular), 
    # p (probabilidad de éxito) 
    # y loc (desplazamiento, en este caso se establece en 0). 
    return result



# Calculate the quantile function from the binomial distribution
'''Esta función usa la función binom.ppf() para calcular el cuantil de una distribución binomial. 
El cuantil es el valor que corresponde a un nivel de probabilidad específico dentro de la distribución binomial. 
El cuantil se calcula a partir del nivel de probabilidad (p), 
el tamaño de muestra (size), 
la probabilidad de éxito (prob) 
y el desplazamiento (loc, que se establece en 0 en este caso)'''

def qbinom(p, size, prob=0.5):
    
    from scipy.stats import binom
    
    result=binom.ppf(q=p,n=size,p=prob,loc=0)
    # La función binom.ppf() devuelve el valor correspondiente al cuantil para el nivel de probabilidad p.
    # Esta función toma varios argumentos: 
    # q (nivel de probabilidad para el cuantil), 
    # n (tamaño de muestra), 
    # p (probabilidad de éxito) 
    # y loc (desplazamiento, en este caso se establece en 0). 
   
    return result



# Generate random variables from the binomial distribution
'''Esta función usa la función binom.rvs() para generar una muestra aleatoria de variables de una distribución binomial.
La muestra aleatoria representa el número de éxitos en una secuencia de ensayos independientes, 
donde cada ensayo tiene dos resultados posibles (éxito o fracaso) 
y la probabilidad de éxito es constante en cada ensayo. 
El tamaño de la muestra aleatoria se determina mediante el parámetro "n",
y el tamaño de muestra en cada ensayo binomial se establece mediante el parámetro "size". '''    

def rbinom(n,size,prob=0.5):
  
    from scipy.stats import binom
    
    result=binom.rvs(n=size,p=prob,size=n)
    # La función binom.rvs() toma varios argumentos: 
    # n (tamaño de la muestra aleatoria a generar), 
    # p (probabilidad de éxito) 
    # y size (tamaño de muestra en cada ensayo binomial). 
    # Devuelve un arreglo de NumPy que contiene la muestra aleatoria generada.
    
    return result
```

```python
dbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) takes the value 2,
pbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) will take a value less than 2
qbinom(0.9, size = 10, prob = 0.2) # what value of a binomial (10,0.2) has a cumulative probability of 0.9?
rbinom(2000, size = 10, prob = 0.2) # generate 2000 random values from a binomial distribution (10,0.2)
```

```python
dbinom(4, size = 18, prob = 0.4)
pbinom(4, size = 10, prob = 0.4)
qbinom(0.5, size = 10, prob = 0.4)
rbinom(2500, size = 10, prob = 0.4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Discrete distributions
xs = np.arange( 1, 11 )
ys = dbinom(xs,size=10,prob=0.5) # compute the shape of the distribution
plt.plot( xs, ys, 'o' )   # plot circles...
plt.vlines( xs, 0, ys )   # ...and lines
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function binomial distribution')
plt.show()
```

Normal Distribution

```python
# Calculate the density of the Normal distribution
'''Esta función utiliza la función norm.pdf() para calcular la función de densidad de probabilidad 
de una distribución normal. La PDF proporciona la densidad de probabilidad en un punto específico "x" 
para una distribución normal con un valor medio "mean" 
y una desviación estándar "sd" dadas.'''

def dnorm(x,mean=0,sd =1):
   
    from scipy.stats import norm


    result=norm.pdf(x,loc=mean,scale=sd)
    # norm.pdf() calcula la PDF de la distribución normal. 
    # Esta función toma varios argumentos: x (punto específico para el cual se desea calcular la PDF), 
    # loc (valor medio) 
    # y scale (desviación estándar). 
    

    return result



# Calculate the cumulative of the normal distribution
'''Esta función utiliza la función norm.cdf() para calcular la función de distribución acumulada 
de una distribución normal. La CDF devuelve la probabilidad acumulada de obtener un valor 
menor o igual a "q" en una distribución normal con un valor medio "mean" y una desviación estándar "sd" dadas.'''

def pnorm(q,mean=0,sd=1):
   
    from scipy.stats import norm


    result=norm.cdf(x=q,loc=mean,scale=sd)
    # La función norm.cdf() calcula la CDF de la distribución normal. 
    # Esta función toma varios argumentos: x (valor hasta el cual se desea calcular la CDF), 
    # loc (valor medio) 
    # y scale (desviación estándar). 


    return result


# Calculate the quantile function of the normal distribution
'''Esta función utiliza la función norm.ppf() de SciPy para calcular el cuantil de una distribución normal. 
El cuantil es el valor que corresponde a un nivel de probabilidad específico dentro de la distribución normal. 
El cuantil se calcula a partir del nivel de probabilidad "p", 
el valor medio "mean" 
y la desviación estándar "sd" de la distribución normal.'''

def qnorm(p,mean=0,sd=1):

    from scipy.stats import norm

    result=norm.ppf(q=p,loc=mean,scale=sd)
    # La función norm.ppf() calcula el cuantil de la distribución normal. 
    # Esta función toma varios argumentos: q (nivel de probabilidad para el cuantil), 
    # loc (valor medio) 
    # y scale (desviación estándar). 

    # La función norm.ppf() devuelve el valor correspondiente al cuantil 
    # para el nivel de probabilidad p en una distribución normal 
    # con el valor medio mean y la desviación estándar sd.


    return result


# Generate random variables from the normal distribution

'''Esta función utiliza la función norm.rvs() de SciPy para generar una muestra aleatoria 
de variables de una distribución normal. 
La muestra aleatoria representa los valores generados a partir de una distribución normal 
con un valor medio "mean" 
y una desviación estándar "sd". 
El tamaño de la muestra aleatoria se determina mediante el parámetro "n".'''
    
def rnorm(n,mean=0,sd=1):
   
    from scipy.stats import norm


    result=norm.rvs(size=n,loc=mean,scale=sd)
    # La función norm.rvs() genera una muestra aleatoria de variables de una distribución normal. 
    # Esta función toma varios argumentos: 
    # size (tamaño de la muestra aleatoria a generar), 
    # loc (valor medio) 
    # y scale (desviación estándar). 
    # Devuelve un arreglo de NumPy que contiene la muestra aleatoria generada.


    return result
```

Uniform distribution

```python
# Calculate the point estimate of the uniform distribution
'''The point estimate for a uniform distribution is simply the average of the minimum and maximum values.'''

from scipy.stats import uniform

def calculate_uniform_point_estimate(min_value, max_value):

    point_estimate = (min_value + max_value) / 2
    return point_estimate


# Calculate the quantile function from the uniform distribution
'''El cuantil es el valor en la distribución uniforme que corresponde a un nivel de probabilidad específico. 
La función toma el nivel de probabilidad deseado, el valor mínimo y el valor máximo de la distribución uniforme 
como argumentos, y devuelve el cuantil correspondiente a ese nivel de probabilidad.'''

from scipy.stats import uniform

def calculate_uniform_quantile(probability, min_value, max_value):
    
    distribution = uniform(loc=min_value, scale=max_value - min_value)
    
    quantile = distribution.ppf(probability)
    # Utiliza el método ppf() de la instancia distribution 
    # para calcular el cuantil correspondiente al nivel de probabilidad "probability". 
    # El método ppf() devuelve el valor en la distribución uniforme 
    # que corresponde al nivel de probabilidad especificado.
    
    return quantile


# Calculate the quantile function from the uniform distribution
'''Este código define una función que calcula el cuantil de una distribución uniforme 
utilizando una fórmula simple en lugar de utilizar la función uniform del módulo scipy.stats. 
La función toma el nivel de probabilidad deseado, 
el valor mínimo 
y el valor máximo de la distribución uniforme como argumentos, 
y devuelve el cuantil correspondiente a ese nivel de probabilidad utilizando la fórmula de interpolación lineal.'''



def calculate_uniform_quantile(probability, min_value, max_value):

    quantile = probability * (max_value - min_value) + min_value
    # Calcula el cuantil de la distribución uniforme utilizando una fórmula simple: 
    # Multiplica el nivel de probabilidad indicado por la diferencia entre el valor máximo 
    # y el valor mínimo de la distribución
    # y luego suma el valor mínimo. 
    # Esto es equivalente a encontrar el valor en la distribución uniforme 
    # que corresponde al nivel de probabilidad especificado.

    return quantile



# Generate random variables from the uniform distribution
'''Este código define una función que utiliza la función np.random.uniform() de NumPy 
para generar un arreglo de variables aleatorias siguiendo una distribución uniforme. 
El rango de la distribución se define mediante los valores min_value y max_value. 
El tamaño del arreglo de variables aleatorias se especifica con el argumento size. 
La función retorna el arreglo de variables aleatorias generadas.'''



def generate_uniform_random_variables(min_value, max_value, size):


    import numpy as np

    random_variables = np.random.uniform(min_value, max_value, size)
    # La función np.random.uniform() de NumPy genera un arreglo de variables aleatorias
    # a partir de una distribución uniforme. 
    # La función toma como argumentos el valor mínimo (min_value), 
    # el valor máximo (max_value) 
    # y el tamaño del arreglo (size).

    return random_variables
```
